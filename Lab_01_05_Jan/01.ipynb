```python
import pathlib
```

```python
def image_converter(input_file_path, output_file_path):
    image = Image.open(input_file_path)
    image.save(output_file_path)
    original_suffix = pathlib.Path(input_file_path).suffix
    new_suffix = pathlib.Path(output_file_path).suffix
    print(f"Converting {input_file_path} from {original_suffix} to {new_suffix}")
```

```python
image_converter("img.jpg", "img.png")
```

```python
image_converter("img.jpg", "img.btmp") # not works on binary, works only for color images.
```

```python
def image_quality(input_file_path, output_file_path, quality):
    image = Image.open(input_file_path)
    image.save(output_file_path, quality=quality, optimize=True)
```

```python
image_quality("img.jpg", "img_compressed.jpg", quality=10) # this will decrease quality to 10% and reduce quality by 90%
```

```python
im2 = Image.open("img_compressed.jpg")
im2
```

```python
def image_converter(input_file_path, output_file_path, dpi):
    image = Image.open(input_file_path)
    image.save(output_file_path, dpi=dpi)
```

```python
image_converter("img.jpg", "image_dpi.jpg", dpi=(30,30))
```

```python
im3 = Image.open("image_dpi.jpg")
im3
```
