### Changing Pixel Colors

```python
from PIL import Image, ImageColor
import matplotlib.pyplot as plt

image = Image.new("RGB", (150, 150))
image
```

```python
red = ImageColor.getcolor("red", "RGB")
green = ImageColor.getcolor("green", "RGB")
color = red
```

```python
count = 0
for y in range(150):
    for x in range(150):
        if count == 5:
            color = red if color != red else green
            count = 0
        image.putpixel((x, y), color)
        count +=1
```

```python
image.save("lines.png")
```

### Converting to Black and White

```python
from PIL import Image

color_image = Image.open("lines.png")
gray_scale = color_image.convert("L")
gray_scale.save("gray_lines.png")
```

```python
gray_scale
```

```python
color_image = Image.open("img.png")
# L = 8 bit pixels, grayscale
gray_scale = color_image.convert("L") # L stands for Luma Transformation
gray_scale
```

```python
gray_scale.getcolors()
```

```python
len(gray_scale.getcolors())
```

```python
# 1 means  1-bit pixels, black and white, stored with one pixel per byte
gray_scale = color_image.convert("1") # gives noisy image, with "salt and pepper noise"
gray_scale
```

```python
gray_scale.getcolors()
```

```python
gray_scale = color_image.convert("1", dither=0)
gray_scale
```

```python
gray_scale.getcolors()
```

```python
from PIL import Image

color_image = Image.open("img.png")

four_colour_image = color_image.convert("P", palette=Image.ADAPTIVE, colors=10)

try:
    four_colour_image.save("four_color_img.png")
    img = Image.open("four_color_img.png")
    img.show()
except :
    pass
```
